## Gumbel Distribution
The Gumbel distribution is a probability distribution that models the maximum (or minimum) of a large number of random samples from other distributions. It's particularly useful for modeling extreme values like maximum floods, maximum wind speeds, or maximum temperatures.
Key Properties

- It belongs to the family of extreme value distributions
- It has two parameters: a location parameter (μ) and a scale parameter (β)
- The PDF (probability density function) has a distinctive asymmetric shape with a longer right tail
- It's sometimes called the Type I extreme value distribution

## Simple Example
Imagine you're monitoring the maximum daily temperature in your city for a month:

- Day 1: 75°F
- Day 2: 78°F
- Day 3: 72°F
...and so on

If you record the maximum temperature each month for many years, the distribution of these monthly maximums will often follow a Gumbel distribution.

## Practical Application
Let's say you're an engineer designing a bridge over a river. You need to know how high to build the bridge to withstand the maximum annual flood level. If you analyze historical flood data, the maximum annual flood heights might follow a Gumbel distribution, which you can then use to estimate the probability of extreme flood events that might only occur once every 50 or 100 years.


## PDF 
$$
f(x; \mu, \beta) = \frac{1}{\beta} \exp\left(-\frac{x-\mu}{\beta} - \exp\left(-\frac{x-\mu}{\beta}\right)\right)
$$

## CDF
$$
F(x; \mu, \beta) = \exp\left(-\exp\left(-\frac{x-\mu}{\beta}\right)\right)
$$

```python
import torch
import matplotlib.pyplot as plt
plt.style.use('bmh')
%matplotlib inline
```

```python
x = torch.distributions.Gumbel(loc=0.0, scale=1.0)
samples = x.sample((2000,))
plt.figure(figsize=(4, 3))
x_range = torch.linspace(-10, 10, 1000)
plt.plot(x_range, x.log_prob(x_range).exp(), label='Gumbel PDF')
plt.hist(samples, bins=100, density=True)
plt.legend()
plt.show()
```
