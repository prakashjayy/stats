## Example 

In a SAT exam, students scores are normally distributed with **$\mu$ = 1100** and  **$\sigma$ = 200**. Ann got 1300 in is SAT score. How well did she perform compared to others in the exam? 

First lets visualize the mean and where Ann score is within the distribution

```python
x = np.random.normal(loc=1100, scale=200, size=1000)
plt.figure(figsize=(8, 3.5))
plt.hist(x, bins=30)
plt.vlines(x=1300, ymin=0, ymax=100, color="red")
plt.vlines(x=1100, ymin=0, ymax=100, color="yellow")
plt.show()
```

```python
from scipy.stats import norm
import math
```

```python
prob_score = norm(1100, 200).cdf(1300)
print(f"Ann performed : {prob_score} compared to other test takers")
```

## Geometric Distribution
then the probability of finding the first success in the nth trial is given by

$$ 
(1 − p)^{n−1} * p
$$

$$
mean (\mu) = 1/p  
$$

$$ 
sd (\sigma) = \sqrt{1-p/p^{2}}
$$

```python
n = 10 
p = 0.7 

x = np.arange(1, 11)
y = [math.pow(1-p, n-1)* p for n in x]
```

```python
plt.figure(figsize=(8, 3.5))
plt.bar(x, y)
plt.title("Geometric distribution")
plt.show()
```
